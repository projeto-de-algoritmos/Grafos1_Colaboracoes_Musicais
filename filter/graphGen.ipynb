# Importando bibliotecas e lendo CSV

```python
import pandas
import numpy
import re
```

```python
data = pandas.read_csv("filtered.csv")
data
```

# Isolando colunas

```python
data = data.loc[:, data.columns.isin(['name', 'artists'])]
data
```

# Removendo caracteres e separando array de artistas

```python
for index, row in data.iterrows():
    row["artists"] = re.sub("( [\[\]']+)|([\[\]']+)", '', row["artists"]).split(",")
```

# Criando array de 'música|artista'

```python
musicas = []
for index, row in data.iterrows():
    for artist in row["artists"]:
        artista = artist
        musica = row["name"]
        musicas.append(f'{musica}|{artista}')

musicas
```

```python
df = pandas.DataFrame(musicas)
df.to_csv("grafos.txt", header=False, index=False)
```
