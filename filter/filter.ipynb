# Importando pandas

```python
import pandas as pd
```

# Lendo CSV

A database escolhida está disponível neste [link](https://www.kaggle.com/datasets/rodolfofigueroa/spotify-12m-songs).
Ela não será anexada ao github por ser muito extensa.

```python
data = pd.read_csv("tracks_features.csv")
data
```

# Separando as colunas

```python
df = data[['name', 'artists', 'release_date']]
df
```

# Filtrando os feats mais recentes

O critério utiliazado para identificar um feat é a presença do caracter ',' que indica que o array possui mais de uma posição e consequentemente mais de um artista.

```python
feat = df[df['artists'].str.contains(",") & (df['release_date'] > '2010')]
feat
```

# Exportando para csv

```python
feat.to_csv("src\data\filtered.csv", index=False)
```
